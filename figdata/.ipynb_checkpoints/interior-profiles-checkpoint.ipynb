# Interior pressure-temperature profiles

## Setup

```julia
using Ogre
using WaterData
using DataFrames
```

## Water planet profiles

We produce pressure-temperature profiles for spheres of water at different surface pressures and temperatures.

```julia
# we need our water EOS and heat capacity
h2o = WaterData.load_full_eos()["grid"]
heatcap = Ogre.GridHeatCapacity("$(Ogre.config.datadir)/tabulated/heatcap-h2o.dat")

# planet boundary conditions setup
Psurf = 5e7
Tsurfs = [300, 500, 1000]
M = 3M_earth
Rguess = R_earth
Rbracket = [0, 10] * Rguess
Npoints = 1000  # high internal resolution

"Make a pressure-temperature profile for a planet"
function PT_profile(Tsurf)
    boundary_vals = Ogre.ValueSet(M, Rguess, Psurf, Tsurf)
    grid = linspace(M, 0, Npoints)
    planet = Ogre.PlanetSystem(M, h2o, heatcap, boundary_vals, 
        grid, Rbracket)
    
    soln = Ogre.find_structure!(planet)
end

# plot to see if it looks reasonable
Ogre.phaseplot(PT_profile(300)) 
```

```julia
# Export presure-temperature profiles

"Make and save a pressure-temperature profile"
function makePT!(Tsurf)
    soln = PT_profile(Tsurf)
    P = vec(Ogre.pressure(soln))
    T = vec(Ogre.temperature(soln))
    df = DataFrame(pressure=P, temperature=T)
    writetable("interior/$(Tsurf)K.csv", df)
end

makePT!(300)
makePT!(500)
makePT!(800);
```
